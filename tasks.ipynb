#### Ordering of Results.

https://docs.python.org/3/library/itertools.html#itertools.combinations

```python
# Label Cups
labels = list(range(total_cups))

#Show
labels
```

```python
# Show the different ways of selecting no_cups_milk_first out of the total number of cups of tea.
combs = list(itertools.combinations(labels, no_cups_milk_first))

#Show
combs
```

```python
# Number of combinations.
len(combs)
```

```python
# Select six cups at random to put milk in first.
# https://docs.python.org/3/library/random.html#random.sample

labels_milk_first = random.sample(labels, 6)

# Sort, inplace.
labels_milk_first.sort()

# Show.
labels_milk_first
```

```python
# Turn labels_milk_first into a set.

# Uses: https://docs.python.org/3/tutorial/datastructures.html#sets
set(labels_milk_first)
```

```python
# Calculate the overlap between each element of combs and labels_milk_first.

no_overlaps = []

for comb in combs:
  # Turn comb into a set.
  s1 = set(comb)

  # Turn labels_milk_first into a set.
  s2 = set(labels_milk_first)


  # Figure out where they overlap.
  overlap = s1.intersection(s2)

  # Show the combination and the overlap.
  print(comb, overlap, len(overlap))
  
  # Append overlap to no_overlaps.
  no_overlaps.append(len(overlap))
```

```python
# Count the number of times each overlap occurs.
counts = np.unique(no_overlaps, return_counts=True)

# Show.
counts
```
