#### 1.3 Ordering of Results.

https://docs.python.org/3/library/itertools.html#itertools.combinations

```python
# Label Cups
labels = list(range(total_cups))

#Show
labels
```

```python
# Show the different ways of selecting no_cups_milk_first out of the total number of cups of tea.
combs = list(itertools.combinations(labels, no_cups_milk_first))

#Show
combs
```

```python
# Number of combinations.
len(combs)
```

Choose cups randomly.

https://docs.python.org/3/library/random.html#random.sample


```python
# Select six cups at random to put milk in first
labels_milk_first = random.sample(labels, 6)

# Sort, inplace.
labels_milk_first.sort()

# Show.
labels_milk_first
```

 Make sets.
 
 https://docs.python.org/3/tutorial/datastructures.html#sets

```python
# Turn labels_milk_first into a set.
set(labels_milk_first)
```

```python
# Calculate the overlap between each element of combs and labels_milk_first.
no_overlaps = []

for comb in combs:
  # Turn comb into a set.
  s1 = set(comb)

  # Turn labels_milk_first into a set.
  s2 = set(labels_milk_first)


  # Figure out where they overlap.
  overlap = s1.intersection(s2)

  # Show the combination and the overlap.
  print(comb, overlap, len(overlap))
  
  # Append overlap to no_overlaps.
  no_overlaps.append(len(overlap))
```

```python
# Count the number of times each overlap occurs.
counts = np.unique(no_overlaps, return_counts=True)

# Show.
counts
```

### 2. numpy's Normal Distribution

**random.normal(loc=0.0, scale=1.0, size=None)** - function used to draw random samples from a normal (Gaussian) 
distribution. It is useful for generating data that follows a normal distribution.

https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html


```python
# Imports

# Numerical structures and operations.
# https://numpy.org/doc/stable/reference/index.html#reference
import numpy as np

# Statistics

import scipy.stats as stats
# https://docs.scipy.org/doc/scipy/reference/stats.html#statistical-functions-scipy-stats

# Plotting

import matplotlib.pyplot as plt
# https://matplotlib.org/stable/api/pyplot_summary.html#module-matplotlib.pyplot

# Statistical models

import statsmodels as sm
# https://www.statsmodels.org/stable/generated/statsmodels.graphics.gofplots.qqplot.html
```

Generate a sample of one hundred thousand values using the function with mean 10.0 and standard deviation 3.0.

```python
# Generate random normal values (mean 10, std 3, 100 000)
data = np.random.normal(10, 3, 100000)

# Show data
data
```

### 2.1 Shapiro-Wilk Test

***scipy.stats.shapiro()*** - function that tests the null hypothesis that the data was drawn from a normal 
distribution.

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.shapiro.html#shapiro


```python
# Perform the Shapiro Wilk test.
stats.shapiro(data)
```

Results

The pvalue is greater than 0.05, there is no reason to suggest that the data was not drawn from normal distribution.
