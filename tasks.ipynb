### Ordering of Results.

**itertools.combinations(iterable, r):** return r length subsequences of elements from the input iterable.

https://docs.python.org/3/library/itertools.html#itertools.combinations

Label cups.

```python
# Label Cups
labels = list(range(total_cups))

#Show
labels
```

### Combinations

```python
# Show the different ways of selecting no_cups_milk_first out of the total number of cups of tea.
combs = list(itertools.combinations(labels, no_cups_milk_first))

#Show
combs
```

```python
# Number of combinations.
len(combs)
```

Choose cups randomly.

**random.sample (population, k, , counts=None):** - return a k length list of unique elements chosen from the population sequence.

https://docs.python.org/3/library/random.html#random.sample


```python
# Select six cups at random to put milk in first
labels_milk_first = random.sample(labels, 6)

# Sort, inplace.
labels_milk_first.sort()

# Show.
labels_milk_first
```

### Make sets.

**Set:**- unordered collection with no duplicate elements.
 
 https://docs.python.org/3/tutorial/datastructures.html#sets

```python
# Turn labels_milk_first into a set.

set(labels_milk_first)
```

Overlap between elements.

```python
# Calculate the overlap between each element of combs and labels_milk_first.
no_overlaps = []

for comb in combs:
  # Turn comb into a set.
  s1 = set(comb)

  # Turn labels_milk_first into a set.
  s2 = set(labels_milk_first)


  # Figure out where they overlap.
  overlap = s1.intersection(s2)

  # Show the combination and the overlap.
  print(comb, overlap, len(overlap))
  
  # Append overlap to no_overlaps.
  no_overlaps.append(len(overlap))
```

Number of overlaps.

```python
# Count the number of times each overlap occurs.
counts = np.unique(no_overlaps, return_counts=True)

# Show.
counts
```

In contrast with the null hypothesis, an alternative hypothesis is developed, which claims that a relationship does exist between two variables.

https://en.wikipedia.org/wiki/Null_hypothesis

### Shapiro-Wilk Test

- a hypothesis test that tests the null hypothesis that a sample comes from a normal distribution. 

- if the p-value is less than the significance level(0.05), we reject the null hypothesis and accept that the sample data is not normally distributed.

- ***scipy.stats.shapiro()*** - function that tests the null hypothesis that the data was drawn from a normal 
distribution.

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.shapiro.html#shapiro


```python
# Perform the Shapiro Wilk test.
stats.shapiro(data)
```

Results

### t-Test Calculation
- test for the null hypothesis that two related or repeated samples have identical average (expected) values.

- **scipy.stats.ttest_rel**:function that calculates the t-test on two related samples of scores, a and b.

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_rel.html#ttest-rel

Libraries

```python
# Numerical arrays
import numpy as np

# Statistical functions
import scipy.stats as stats
```

Calculating t-statistic

Dataset:

Before: [63, 68, 70, 64, 74, 67, 70, 57, 66, 65]

After: [64, 64, 68, 64, 73, 70, 72, 54, 61, 63]

```python
# Pre-exercise heart rate.
before = np.array([63, 68, 70, 64, 74, 67, 70, 57, 66, 65])

# Post-exercise heart rate.
after = np.array([64, 64, 68, 64, 73, 70, 72, 54, 61, 65])
```

```python
# Calculate the differences 
differences = after - before

# Calculate the mean of the differences
mean_diff = np.mean(differences)
```

```python
# Calculate the standard deviation of the differences
std_diff = np.std(differences, ddof=1)  # ddof=1 for sample standard deviation

# Number of pairs
n = len(differences)
```

```python
# Calculate the t-statistic
t_stat = mean_diff / (std_diff / np.sqrt(n))
print(t_stat)
```

#### Scipy.stats.ttest_rel
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_rel.html#ttest-rel

```python
# Paired sample t-test
stats.ttest_rel(before, after)
```

The t-statistic calculated manually and the one obtained from scipy.stats.ttest_rel match, showing that both methods yield consistent results.

The p-value is approximately 0.302, higher than the significant level (0.05).
Based on this data, there is no statistically significant difference between the heart rates before and after the exercise program


### 4. ANOVA

**Analysis of variance (ANOVA)** is a statistical test used to evaluate the difference between the means of more than two groups (whether differences are meaningful or due to random chance).

https://www.investopedia.com/terms/a/anova.asp

https://statistics.laerd.com/spss-tutorials/one-way-anova-using-spss-statistics.php



The null hypothesis : All group means are equal.

The alternative hypothesis : At least one group mean is different.


**Type II Error:**

An error that occurs when one fails to reject a null hypothesis that is actually false. A type II error produces a false negative, also known as an error of omission.

https://www.investopedia.com/terms/t/type-ii-error.asp

**One way Anova:** 

Tests the null hypothesis that two or more groups have the same population mean. The test is applied to samples from two or more groups, possibly with differing sizes.

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.f_oneway.html#f-oneway


```python
import numpy as np # for numerical arrays
import scipy.stats as stats # statistical functions
from scipy.stats import f_oneway # statistical functions
```

Three samples (each with 100 values)

Sample 1: Mean = 4.9, Std Dev = 0.1

Sample 2: Mean = 5.0, Std Dev = 0.1

Sample 3: Mean = 5.1, Std Dev = 0.1

Number of simulations = 10 000


```python
# Perform 10 000 simulation
def simulate_ttest(alpha=0.05):

    # Generate three samples.
    sample1 = np.random.normal(loc=4.9, scale=0.1, size=100)
    sample2 = np.random.normal(loc=5.0, scale=0.1, size=100)
    sample3 = np.random.normal(loc=5.1, scale=0.1, size=100)

# Perform one way ANOVA to test if ther's a significant difference among the means.

    #_, p_value = stats.f_oneway(sample1, sample2, sample3)
    f_statistic, p_value = f_oneway(sample1, sample2, sample3)
# Check for Type II error (p-value > 0.05)
    if p_value > alpha:
        return True
    else:
        return False
```

```python
simulate_ttest()
```

```python
f_statistic, p_value
```

```python
# Initialize the counter for Type II errors
no_type_ii = 0

# Perform 10,000 simulations.
for _ in range(10000):
    # Count the number of times a Type II error occurs
    if simulate_ttest():
        # Increment the number of errors
        no_type_ii += 1

# Show the number .
no_type_ii
```

### Results

- There's no probability of committing Type II error.
- The p-value is less than 0.05 therefore there's no evidence to accept the NULL hypothesis.
